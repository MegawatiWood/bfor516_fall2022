## Convert Dates
We need to convert the dates in the data file to `datetime` format. They are currently stored as
strings. Fortunately, `pandas` includes a function to perform this task for us.

```python
# convert dates
weather['timestamp'] = pd.to_datetime(weather['timestamp'])
```

# Column summaries
We can use the `describe()` function in `pandas` to see basic information about our dataframe.

```python
weather.describe(include='all', datetime_is_numeric=True)
```

# Lab
For this week's lab, you will summarize the data in a few different ways.

First, we can get the annual average temp. This 
[StackOverflow Question](https://stackoverflow.com/questions/24082784/pandas-dataframe-groupby-datetime-month)
explains what is happening. For even more detail, you can also look at the 
[`pandas` documentation for the `Grouper` function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Grouper.html?highlight=grouper).




```python
# group by month
annual_avg_tmp = weather.groupby(pd.Grouper(key='timestamp', freq='Y')).mean()
annual_avg_tmp
```

## Monthly summary
We can also get the average temperature for each month in the dataset.

```python
monthly_avg_tmp = weather.groupby(pd.Grouper(key='timestamp', freq='M')).mean()
monthly_avg_tmp
```

## Visualizing
Plot the monthly averages over time.

```python
# monthly_avg_tmp['mean_tmp'].plot()
weather['mean_tmp'].plot()
```
